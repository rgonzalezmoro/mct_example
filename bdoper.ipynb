```python
# Import librerías python
import sqlite3
import sys
import os
from datetime import datetime

# Import de modulos ipynb
import import_ipynb
import environ

bbdd_path = os.getenv('BBDD_PATH')

def upadate_sign(id, fake_signed):
    try:
        _sqlConn = sqlite3.connect(str(bbdd_path))
        _cursor = _sqlConn.cursor()
    except:
        sys.exit('Error al conectar a la base de datos en upadateSignedHash')
    
    try:
        update_query = "UPDATE tableusers set signed_hash = ? where id = ?"
        data = (fake_signed, id)
        _cursor.execute(update_query, data)
        _sqlConn.commit()
        print(f'Falseada la firma del hash de {id}')
    except:
        sys.exit('Error en UPDATE tableusers falseando firma')
    finally:
        if (_sqlConn):
            _sqlConn.close()

def insert_user(_nombre, _clave_rsa, _serialized_hash, _signed_hash):
    try:
        _sqlConn = sqlite3.connect(str(bbdd_path))
        _cursor = _sqlConn.cursor()
    except:
        sys.exit('Error al conectar a la base de datos en insert_user')

    _timestamp = datetime.utcnow().strftime("%d/%m/%Y, %H:%M:%S")
    _list_values = [_nombre, _clave_rsa, _serialized_hash,_signed_hash, _timestamp]

    try: 
        _insert_query = "INSERT INTO tableusers (nombre, clave_rsa, hash, signed_hash, timestamp) \
                        VALUES (?,?,?,?,?)"
        _cursor.execute(_insert_query, _list_values)
        _sqlConn.commit()
        
        print(f'{_nombre} añadido a la tabla')

    except sqlite3.Error as error:
        sys.exit("Error al insertar registro en tableusers - {} ".format(error))
    finally:
        if (_sqlConn):
            _sqlConn.close()
                       
def select_user(_id):
    try:
        _sqlConn = sqlite3.connect(str(bbdd_path))
        _cursor = _sqlConn.cursor()
    except:
        sys.exit('Error al conectar a la base de datos en select_user')
        
    try:
        select_query = 'SELECT * FROM tableusers WHERE id = ?'
        _cursor.execute(select_query, (_id, )) 
        record = _cursor.fetchone()
        return record
    except:
        sys.exit('Error al seleccionar el ultimo registro')
    finally:
        if (_sqlConn):
            _sqlConn.close()

def select_all_user():
    try:
        _sqlConn = sqlite3.connect(str(bbdd_path))
        _cursor = _sqlConn.cursor()
    except:
        sys.exit('Error al conectar a la base de datos en select_user')
        
    try:
        select_query = 'SELECT * FROM tableusers'
        _cursor.execute(select_query) 
        record = _cursor.fetchall()
        return record
    except:
        sys.exit('Error al seleccionar el ultimo registro')
    finally:
        if (_sqlConn):
            _sqlConn.close()
            
def deleteTable(_name):
    try:
        _sqlConn = sqlite3.connect(str(bbdd_path))
        _cursor = _sqlConn.cursor()
    except:
        sys.exit('Error al conectar a la base de datos en deleteTable')
    
    try: 
        _delete_query = "delete from " + _name
        _cursor.execute(_delete_query)
        _sqlConn.commit()
    except sqlite3.Error as error:
        sys.exit("Error al borrar tabla {} - {} ".format(_name,error))
    finally:
        if (_sqlConn):
            _sqlConn.close()
```
